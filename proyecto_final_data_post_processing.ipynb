```python
import pandas as pd
import numpy as np

```

```python
path_obs = '/content/df_obs_final_20231114.csv'
path_fcast = '/content/df_fcast_final_20231114.csv'
```

```python
fcst = pd.read_csv(path_fcast)
obs = pd.read_csv(path_obs)

fcst.head()
```

```python
obs.head()
```

```python
fcst.rename(columns = {'Unnamed: 0' : 'date'}, inplace = True)
obs.rename(columns = {'Unnamed: 0' : 'date'}, inplace = True)
```

```python
fcst['tipo'] = 'fcst'
obs['tipo'] = 'obs'
```

```python
data0 = pd.concat([obs, fcst], axis =0 )
data0.sort_values(by='date', inplace=True)
data0.reset_index(drop=True, inplace=True)
print(data0.shape)
data0.iloc[151:158,:]
```

```python
data1 = pd.merge(obs, fcst, on='date', how='outer', suffixes = ['_obs', '_fcst'])
data1.sort_values(by='date', inplace=True)
data1.reset_index(drop=True, inplace=True)
print(data1.shape)
data1.loc[151:158,['company_obs', 'company_fcst']]
```

```python
data1.to_csv('data1.csv')
data0.to_csv('data0.csv')
```
