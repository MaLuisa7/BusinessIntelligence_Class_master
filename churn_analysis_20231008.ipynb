```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/EarlyRetirementPrediction (1).xlsx")
df = df.iloc[:, :-1]
df.head()
```

```python
df.shape
```

```python
df.columns = ['Retire_before_65', 'Gender', 'Desease', 'Education_level',
       'Marital_status', 'Monthly_income', 'Employee_employer', 'Dependants',
       'Unemployment_rate', 'Stock_market', 'Credit_score',
       'Government_bonds_return' ]
```

```python
df.describe()
```

```python
df.Retire_before_65.value_counts()
```

```python
df.Retire_before_65.value_counts()/len(df)
```

```python
X = df.drop(columns = ["Retire_before_65"])
y = df.Retire_before_65
```

```python
X_norm = (X - X.min()) / (X.max() - X.min())
X_norm.head()
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split( X_norm, y, test_size=0.30, random_state=0)
```

```python
ytrain.value_counts() / len(ytrain)
```

```python
ytest.value_counts() / len(ytest)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 1)
classifier.fit(xtrain, ytrain)
y_pred = classifier.predict(xtest)
```

```python
res_test = pd.DataFrame(data = classifier.predict_proba(xtest), columns = ['pred_0', 'pred_1'])
res_test['clase'] = ytest.reset_index(drop=True)

res_test['pred'] = y_pred
res_test.head()
```

```python
res_test.to_csv('resultados_test_20231008.csv')
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(ytest, y_pred)
print ("Confusion Matrix : \n", cm)
```

```python
from sklearn.metrics import accuracy_score , recall_score, precision_score

print ("Accuracy : ", accuracy_score(ytest, y_pred))
print ("recall : ", recall_score(ytest, y_pred))
print ("precision : ", precision_score(ytest, y_pred))
```
